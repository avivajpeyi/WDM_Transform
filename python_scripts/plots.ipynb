```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Plot chirp


```python
fname = '../chrp_time.dat'
data = np.loadtxt(fname)
plt.plot(data[:,0], data[:,1])
plt.xlabel('Time (s)');
plt.ylabel('Amplitude');
```

```python
# FFT and plot spectogram
from scipy.fftpack import fft
from scipy.signal import spectrogram
fs = 1/(data[1,0]-data[0,0])
f, t, Sxx = spectrogram(data[:,1], fs, nperseg=1024)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
def read_xyz_data(fname):
    data = np.loadtxt(fname)
    # reshape Z array to 2D based on X and Y arrays
    z = data[:,2]
    x = np.unique(data[:,0])
    y = np.unique(data[:,1])
    z = z.reshape(len(x), len(y))
    return x, y, z

def plot_xyz_imshow(x, y, z):
    plt.imshow(z, extent=(x.min(), x.max(), y.min(), y.max()), aspect='auto', origin='lower')
    plt.colorbar()
    plt.xlabel('Time (s)');
    plt.ylabel('Frequency (Hz)');
    return plt.gcf()

fname = '../BinaryF.dat'
fig = plot_xyz_imshow(*read_xyz_data(fname))
fig.axes[0].set_ylim(0, 200)
fig.show()
```
